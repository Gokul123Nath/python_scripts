```python
# Plot (Usual line plot)
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('_mpl-gallery')
x = np.linspace(0,1,10)
y = 4+2 * np.sin(2*x)

plt.plot(x,y, linewidth=2.0, linestyle  = '-')

# In order to set something like xticks, xlim you have to go with subplot

# fig,ax = plt.subplots()
# ax.plot(x,y)
# ax.set(xlim=(0,9), xticks=np.arange(1,8))
plt.legend(['line Plt'])

plt.show()
```

```python
# Scatter plot along with color (c - parameter)

x=[1,2,3,4,5,6]
y = [3,4,9,5,4,1]

plt.scatter(x,y, c='black')
plt.xlabel('fin')
plt.title('Gokl')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

# plot:
fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.show()
```

```python
import numpy as np

# Coefficient matrix A
A = np.array([[1, 2], [3, 5]])

# Right-hand side values (constants)
b = np.array([1, 2])

# Solve for x
x = np.linalg.solve(A, b)

print("Solution vector x:", x)
```
